```python
import pandas as pd
df_corpus = pd.read_csv("/Users/andrea/Desktop/US_youtube_trending_data.csv")
```

```python
df_corpus.head()
```

```python
import texthero as hero
```

```python
df_corpus['clean_tags'] = hero.clean(df_corpus['tags'])
```

```python
df_corpus['clean_tags'].head
```

```python
import matplotlib.pyplot as plt
NUM_TOP_WORDS = 20
top_20 = hero.visualization.top_words(df_corpus.loc[:, 'clean_tags']).head(NUM_TOP_WORDS)
top_20.plot.bar(rot=90, title="Top 20 tags in most popular videos");
plt.show();
```

```python
hero.wordcloud(df_corpus.loc[:, 'clean_tags'], max_words=100)
```
